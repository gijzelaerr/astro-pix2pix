```python
import os
import matplotlib.pyplot as plt
from astropy.io import fits
%matplotlib inline
```

```python
path = "."
```

```python
def open_fits(f):
    return fits.open(f)[0].data.squeeze()
```

```python
for i in [j for j in os.listdir(path) if j.endswith('.fits')]:
    data = open_fits(os.path.join(path, i))
    f, (a) = plt.subplots(1, 1, figsize=(6,6))
    a.pcolor(data, cmap='gist_heat')
    a.set_xticks([])
    a.set_yticks([])
    plt.savefig(os.path.join(path, i[:-5] + ".png"), bbox_inches='tight')
```
