```python
from astropy.io import fits
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
target = (256, 256)
```

```python
psf_big = fits.open('/home/gijs/examples/ddf/psf-big.fits')[0].data.squeeze()
psf_big = psf_big / psf_big.max()
```

```python
psf_small_ref = fits.open('/home/gijs/examples/ddf/wsclean-psf.fits')[0].data.squeeze()
```

```python
psf_small = psf_big[psf_big.shape[0]//2-target[0]//2+1:psf_big.shape[0]//2+target[0]//2+1,
        psf_big.shape[1]//2-target[1]//2+1:psf_big.shape[1]//2+target[1]//2+1]
(psf_small - psf_small_ref).max()
```
